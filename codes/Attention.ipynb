```python
from konlpy.tag import Kkma as km
from konlpy.tag import Okt
from sklearn.feature_extraction.text import TfidfVectorizer
import re
```

```python
def read_data(filename):
    with open(filename, 'r', encoding="utf-8") as f:
        datas = [doc for doc in str(f.read()).split('\n')]
        # txt 파일의 헤더(id document label)는 제외하기
    return datas
```

```python
datas = read_data('../datas/news_and_sums.txt')
sums = read_data('../datas/sum2.txt')
caution= read_data('../datas/caution.txt')
response = read_data('../datas/response.txt')
```

```python
print(datas)
datas = datas[0].split('.')
last_data = datas[:]
for data in datas:
    print(data)
    print('\n')
print(datas)
```

```python
def reviewTokenize(data):
    data = str(data).replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]", "")
    data = re.sub('[-=+,#/\?:^$.@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>▲�`\'…》]', '', data)
    return data
```

```python
def tokenize(doc):
    # norm은 정규화, stem은 근어로 표시하기를 나타냄
    okt = Okt()
    docs =[]
    for t in okt.pos(doc, norm=True, stem=True):
        if(t[1] != "Punctuation" and t[1] != "Foreign" and t[1] != "Josa" and t[1] != "Email"):
            docs.append('/'.join(t))
        print(t)
        
    return docs
```

```python
for index, data in enumerate(datas):
    datas[index] = tokenize(data)
print(datas)
```

```python
print(datas)
print(sums)
```

```python
final = []
final_caution= []
final_response = []
for idx, data in enumerate(datas):
    final.append(' '.join(data))
    final_caution.append(' '.join(data))
    final_response.append(' '.join(data))
print(final)
final.append(' '.join(sums))
final_caution.append(' '.join(caution))
final_response.append(' '.join(response))
print(len(datas))
```

```python
tfidf_vectorizer = TfidfVectorizer(min_df=1)
tfidf_matrix_kkma = tfidf_vectorizer.fit_transform(final)
tfidf_matrix_kkma2 = tfidf_vectorizer.fit_transform(final_caution)
tfidf_matrix_kkma3 = tfidf_vectorizer.fit_transform(final_response)
document_distances = (tfidf_matrix_kkma * tfidf_matrix_kkma.T)
document_distances2 = (tfidf_matrix_kkma2 * tfidf_matrix_kkma2.T)
document_distances3 = (tfidf_matrix_kkma3 * tfidf_matrix_kkma3.T)
```

```python
arr = document_distances.toarray()
arr2 = document_distances2.toarray()
arr3 = document_distances3.toarray()
print(len(arr[-1]))
print("\n피해단계")
for index, pre in enumerate(arr[-1]):
    if pre *100 > 4 and index != len(arr[-1])-1:
        print(last_data[index])
print('\n유의사항')
for index, pre in enumerate(arr2[-1]):
    if pre *100 > 4 and index != len(arr2[-1])-1:
        print(last_data[index])
print('\n피해대응')
for index, pre in enumerate(arr3[-1]):
    if pre *100 > 4 and index != len(arr3[-1])-1:
        print(last_data[index])
```
