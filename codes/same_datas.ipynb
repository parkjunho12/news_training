```python
from konlpy.tag import Kkma as km
from konlpy.tag import Okt
from sklearn.feature_extraction.text import TfidfVectorizer
import re
import copy
```

```python
def read_data(filename):
    with open(filename, 'r', encoding="utf-8") as f:
        datas = [doc for doc in str(f.read()).split('다음 기사에요')]
        # txt 파일의 헤더(id document label)는 제외하기
    return datas
```

```python
datas = read_data('../datas/origin_file.txt')
sums = read_data('../datas/sums_file.txt')
```

```python
def tokenize(doc):
    # norm은 정규화, stem은 근어로 표시하기를 나타냄
    okt = Okt()
    docs =[]
    for t in okt.pos(doc, norm=True, stem=True):
        if(t[1] != "Punctuation" and t[1] != "Foreign" and t[1] != "Josa" and t[1] != "Email"):
            docs.append('/'.join(t))     
    return docs
```

```python
def remove_space(datas):
    doc = []
    for idx, data in enumerate(datas):
        doc.append(data.split('\n'))
    return doc
```

```python
datas = remove_space(datas)
over_data = copy.deepcopy(datas)
print(over_data)
```

```python
def vectorize_data(docs, sum_data):
    last_data = []
    for doc in docs:
        last_data.append(' '.join(doc))
    last_data.append(' '.join(sum_data))
    print(sum_data)
    tfidf_vectorizer = TfidfVectorizer(min_df=1)
    tfidf_matrix = tfidf_vectorizer.fit_transform(last_data)
    distances = (tfidf_matrix * tfidf_matrix.T)
    return distances.toarray()
```

```python
def predict_sentence(array, final):
    lat_data = ""
    for idx, data in enumerate(array[-1]):
        if data * 100 > 80 and len(array[-1]) - 1 != idx:
             print(final[idx])
    print(array[-1])
            
```

```python
for index, data in enumerate(datas):
    for idx, da in enumerate(data):
        data[idx] = tokenize(da)
    sums[index] = tokenize(sums[index])
    last_array = vectorize_data(data, sums[index])
    predict_sentence(last_array, over_data[index])
    print('\n')
```
